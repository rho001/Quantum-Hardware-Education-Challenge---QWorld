# Quantum optics

You may have learnt a small chunk of quantum optics during senior high school. Light is composed of particles called photon. When the energy of the light is large, or when the number of photon is very large, we have our light that behaves classically. But when we decrease the energy of light down to several photons level, we will see that it will have a unique quantum behavior.

$$
E=nhf
$$
where
$$
E=\mathrm{energy\,of\,light}
$$
$$
n=\mathrm{number\,of\,photons}
$$
$$h=\mathrm{Planck's\,constant}
$$

For example, if we send light through 50:50 beam splitter and measure it using light detector, we should see that both detector click because the light is divided equally into both path. However, if you decrease the energy down to a single photon level, you will see that only one of the detector will click at any point of time because the photon can only go through one path. We will see that any one of the detector will click 50% of the time.

Next, we will see more about the formalism of quantum optics.

# Quantum Optics formalism

Maxwell's equations are the foundation of classical electromagnetism. Any function that describes electromagnetic field must satisfy the Maxwell's equations. Given below is the Maxwell's equation for electromagnetic field in vacuum without electric charge. You don't need to know what the equations below mean. What you need to know is that $E$ and $B$ represents the strength (and direction) of the electric and magnetic field.

\begin{aligned}
\nabla \cdot \mathbf{E} &=0 \\
\nabla \cdot \mathbf{B} &=0 \\
\nabla \times \mathbf{E} &=-\frac{\partial \mathbf{B}}{\partial t} \\
\nabla \times \mathbf{B} &=\frac{1}{c^{2}} \frac{\partial \mathbf{E}}{\partial t}
\end{aligned}

The solution to the set of Maxwell's equations above is:

$\mathbf{E}(\mathbf{r}, t)=\vec{\varepsilon}_{\ell} \mathcal{E}_{\ell}(t) \operatorname{e^{i\left\{\mathbf{k}_{\ell} \cdot \mathbf{r}\right\}}} + $ c.c.

where $\vec{\varepsilon}_{\ell}$ is a unit vector describing the direction, $\mathcal{E}_{\ell}(t)$ is the complex amplitude, $\mathbf{k}_{\ell}$ is the wave number and c.c. is a complex conjugate of the first term. Again, you don't need to know what all these symbols mean.

Furthermore, the complex amplitude $\mathcal{E}_{\ell}(t)$ can be further described as a product of the time-dependent component and time-independent component:

$$
\mathcal{E}_{\ell}(t)=\mathcal{E}_{\ell}(0) \exp \left\{-\mathrm{i} \omega_{t} t\right\}
$$

Where $\omega_{\ell}=c k_{\ell}$.

The time-independent component $\mathcal{E}_{\ell}(0)$ plays an important role. The real and imaginary part of it are a pair of cannonically conjugate variable up to some constant factor. We call these two components as the quadratures.

$$
x=\mathrm{Re}\left(\mathcal{E}_{\ell}(0)\right)
$$

$$
p=\mathrm{Im}\left(\mathcal{E}_{\ell}(0)\right)
$$

In quantum mechanics, things are almost looking the same. The electric and magnetic field can be written in a similar fashion (this depends on how deep the Physics that you want to study, but the conclusion is surprisingly almost the same). If you take into account quantum mechanics, the only difference is that now $x$ and $p$ that you see will not be numbers anymore, but a quantum operator. This procedure is called canonical quantization by physicists. We won't go into the details, but we will substitute the quadratures $x$ and $p$ with the quantum operators $\hat{x}$ and $\hat{p}$.

One of the consequence is that the measurement of the quadratures will exhibit Heisenberg uncertainty principle, i.e.,

$$
\sigma_{x} \sigma_{p} \geq \frac{\hbar}{2}
$$

we can also define two new operators $\hat{a}$ and $\hat{a}^{\dagger}$ defined below:

$$
\hat{a}=\frac{\hat{q}+i\hat{p}}{2}
$$

$$
\hat{a}^{\dagger}=\frac{\hat{q}-i\hat{p}}{2}
$$

$\hat{a}$ is called annihilation operator and $\hat{a}^{\dagger}$ is called creation operators. The reason why they are called that way is that when you apply creation operator to "state with no light" (called a vacuum state), it will create a single photon, i.e.,

$$
\hat{a}^{\dagger} | 0 \rangle=| 1 \rangle
$$

where $| 1 \rangle$ is the state respresenting a single photon. As you might be able to guess, the annihilation operator will annihilate (or destroy) a photon. In general,

$$\begin{aligned} a^{\dagger}|n\rangle &=\sqrt{n+1}|n+1\rangle \\ a|n\rangle &=\sqrt{n}|n-1\rangle \end{aligned}$$

where $|n\rangle$ is the state of definite n number of photons. Note that we have dropped the hats for brevity, but always remember that we are dealing with quantum operators. One can also define a photon number operator as follows

$$N =a^{\dagger} a$$

and it turns out that this number state or Fock state is the eigenstate this operator.

$$N|n\rangle =n|n\rangle$$

This means that the operator $N$ measures the photon number.

# States

In photonic quantum computing platform, there are several states that are commonly discussed, either to ease the theoretical analysis or because they are commonly used in practice.

One simple state is the quadrature eigenstates $|x\rangle$ and $|p\rangle$. These states are the states of definite $x$ and $p$ quadrature, respectively. These states are not physical, or in other words, cannot be produced in practice. Thus, these are introduced only for theoretical purpose.

Coherent state is a quantum state that is the closest to classical light, i.e., it exhibits a mininum uncertainty:

$$
\sigma_{x}=\sigma_{p}=\sqrt{\frac{\hbar}{2}}
$$

Coherent state can be described in polar coordinates $|r^{i\theta}\rangle$ or cartesian coordinates $|q+ip\rangle$, with $r^{i\theta}$ typically denoted by $\alpha = r^{i\theta}$. $|\alpha|^2$ represents the energy of the coherent state.

Squeezed state is a coherent state where one of the quadrature is squeezed (decreased variance).

Fock state is a state that has definite number of photons $| n \rangle$.

# CV vs DV

In photonic quantum computing platform, we typically utilize a continuous-variable (CV) eigenstate to encode our information, instead of the discrete eigenstate of the typical discrete-variable (DV) based quantum computer. This eigenstate is usually the X quadrature that we have discussed above. For any state, we can expand it in the CV basis:

$$
|\psi\rangle=\int d x \psi(x)|x\rangle
$$

where $|x\rangle$ is the quadrature state and the value $x$ can take any real value. This basis is called *qumode*. This is in contrast to the DV-based quantum computer that has state expansion as:

$$
|\phi\rangle=\phi_{0}|0\rangle+\phi_{1}|1\rangle
$$

# Gaussian states

There is a class of states that is very important in photonic quantum computer which is called Gaussian state.

## Vacuum state

In Strawberry Fields, we use blackbird porgramming language to describe and synthesize our quantum circuits. You don't need to worry about this yet because using it is quite simple.

```python
#Creating the the photonic quantum circuit that is represented as "program"
#We will only declare the state

prog = sf.Program(1) #declaring a circuit with 1 qumode (similar to qubit)

with prog.context as q:
    Vac | q[0] #declaring a vacuum state in the q[0]
```

Vacuum state is usually denoted by $|0\rangle$. It represents the state of no photon. 

To run the program, we need to specify the hardware. You can either run the program on your local hardware or remote hardware as simulation, or run it on the actual photonic quantum computer device. For this time, we will run it on our local computer. We choose the simulator to use the "Gaussian" backend, this essentially determines what kind of algorithm the simulator uses for the simulation. For now, you don't need to worry about this.

```python
eng = sf.Engine("gaussian") #choosing the simulator to be gaussian

result = eng.run(prog) #run the circuit
state = result.state #probe the quantum state of the output
state.quad_expectation(0) #view the quadrature expectation at angle 0
```

The quad_expectation(0) method will return two numbers, which are the X quadrature mean and X quadrature variance. As you can see from the cell above, the mean is 0 and the variance is 1.

### Quadrature Measurement

There are three measurements available in Strawberry Fields: homodyne, heterodyne and photon number. We will talk about homodyne measurement first.

```python
prog = sf.Program(1)

with prog.context as q:
    Vac | q[0]
    MeasureHomodyne(phi=0) | q[0]
```

```python
#set simulator engine as Gaussian
eng = sf.Engine(backend='gaussian')

#run the engine
results = eng.run(prog)
```

```python
#print the result
print(results.samples[0][0])
```

Now, let's do the measurement for 10000 times and plot the histogram.

```python
n=10000
q=np.zeros(n)
for i in range(0,n):
    result=eng.run(prog)
    q[i]=result.samples[0][0]
```

```python
plt.hist(q,bins=100);
```

The $x$ and $p$ quadratures are the projection of the state $|r e^{i\theta}\rangle$ at the complex plane. But you can also measure the quadrature as a projection with certain angle $\phi$.

```python
prog = sf.Program(1)

with prog.context as q:
    Coherent(r=2,phi=2) | q[0]
    MeasureHomodyne(phi=0.79) | q[0] #random 

n=10000
q=np.zeros(n)
for i in range(0,n):
    result=eng.run(prog)
    q[i]=result.samples[0][0]
```

```python
plt.hist(q,bins=100);
```

You can also view the quadrature mean and variance at arbitrary angle by using quad_expectation(mode, $\phi$), where mode is the index of the register (let's say 1 if you want to look at the q[1] register).

```python
result = eng.run(prog) #run the circuit
state = result.state #probe the quantum state of the output
state.quad_expectation(0,0.79) #view the quadrature expectation at angle 0
```

## Coherent state

Use Coherent(r,phi) to declare a coherent state. This is a typical output state of a laser. Coherent state has two arguments, r, which is the displacement magnitude which represents the strength or energy of the state and phi, which is the phase angle.

Let's denote $\alpha = r^{i\phi}$, then the quadratures of the Coherent state can be written as $X = 2\sqrt{\frac{\hbar}{2}}Re\left(\alpha\right)$, where $Re\left(z\right)$ denotes the real part of the complex number $z$ and $P = 2\sqrt{\frac{\hbar}{2}}Im\left(\alpha\right)$, where $Im\left(z\right)$ denotes the imaginary part of the complex number $z$.

In SF, $\hbar=2$ because of the unit choice of the platform, but this can be changed.

In the example below, we declare a coherent state with $r = 0.5$ and $\phi = \pi/3$.

```python
#Creating the the photonic quantum circuit that is represented as "program"
#We will only declare the state

prog = sf.Program(1) #declaring a circuit with 1 qumode (similar to qubits)

with prog.context as q:
    Coherent(r=0.5, phi=np.pi/3) | q[0] #declaring a coherent state in the q[0]
```

```python
result = eng.run(prog)
state = result.state
state.quad_expectation(0)
```

## Squeezed state

Squeezed state is a vacuum state that has its quadrature variance squeezed along an axis (and quadrature variance spread out along an axis orthogonal to the axis in which it is squeezed). We will learn about this later on. However, you can see what squeezing means by looking at the following example.

```python
#Creating the the photonic quantum circuit that is represented as "program"
#We will only declare the state

prog = sf.Program(1) #declaring a circuit with 1 qumode (similar to qubits)

with prog.context as q:
    Squeezed(r=1, p=0) | q[0] #declaring a squeezed state in the q[0]
```

```python
result = eng.run(prog)
state = result.state
state.quad_expectation(0)
```

## Displaced squeezed state

A displaced squeezed state will have four parameters
r_d: magnitude of displacement,
phi_d: direction of displacement,
r_s: magnitude of squeezing,
phi_s: direction of squeezing

In blackbird programming language, it is implemented as
DisplacedSqueezed(r_d, phi_d, r_s, phi_s) | register

```python
#Creating the the photonic quantum circuit that is represented as "program"
#We will only declare the state

prog = sf.Program(1) #declaring a circuit with 1 qumode (similar to qubits)

with prog.context as q:
    DisplacedSqueezed(r_d=1.0, phi_d=0.0, r_s=1.0, phi_s=0.0) | q[0] #declaring a squeezed state in the q[0]
```

```python
result = eng.run(prog)
state = result.state
state.quad_expectation(0)
```

As you can see the X quadrature is displaced (as the mean is nonzero) and squeezed (as the variance is less than 1 unit).
